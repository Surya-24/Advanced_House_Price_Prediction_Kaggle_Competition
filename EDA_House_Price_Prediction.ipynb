```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#display all columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
data = pd.read_csv('train.csv')
```

```python
print(data.shape)
data.head()
```

#Data Analysis

```python
#Checking Missing Values
features_with_na = [features for features in data.columns if data[features].isna().sum()>1]
```

```python
#Getting the percentage of null values
for feature in features_with_na:
  print(feature,np.round(data[feature].isna().mean(),4), '%missing values')
```

Since they are many missing values, we need to find the relationship between missing values and Sales Price

```python
for feature in features_with_na:
    dat = data.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    dat[feature] = np.where(dat[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    dat.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here there is relationship with tht null values so we need to replace those null values with something useful

#Numerical Values

```python
numerical_features = [feature for feature in data.columns if data[feature].dtypes != 'O'] #O is object type
print(f"No. of Numerical Columns ={len(numerical_features)}")
```

```python
data[numerical_features].head()
```

##Temporal Variables(Datetime Variables) <br>
From the Dataset we have 4 year variables. We have extract information from the datetime variables like no of years or no of days. One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold.

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
#Let's see what all years we have
for year in year_feature:
  print(year, data[year].unique()) 
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price

data.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
## Here we will compare the difference between All years feature with SalePrice

for feature in year_feature:
    if feature!='YrSold':
        dat=data.copy()
        ## We will capture the difference between year variable and year the house was sold for
        dat[feature]=dat['YrSold']-dat[feature]

        plt.scatter(dat[feature],dat['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

Discrete Variables

```python
discrete_features = [feature for feature in numerical_features if(len(data[feature].unique())<25 and feature not in year_feature+['Id'])]
discrete_features
```

```python
print("Discrete Variables Count: {}".format(len(discrete_features)))
```

```python
data[discrete_features].head()
```

```python
for feature in discrete_features:
    dat=data.copy()
    dat.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature = [feature for feature in numerical_features if feature not in discrete_features+year_feature+['Id']]
continuous_feature
```

```python
for feature in continuous_feature:
    dat=data.copy()
    dat[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

These have skewed data... which we need to convert

```python
# We will be using logrithmic transformation on our data
for feature in continuous_feature:
  dat = data.copy()
  if 0 in dat[feature].unique():
    pass
  else:
    dat[feature] = np.log(dat[feature])
    dat['SalePrice'] = np.log(dat['SalePrice'])
    plt.scatter(dat[feature],dat['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
dat['SalePrice'].head()
```

We can see some correlation b/w our features and Sales Price after applying logrithmic Distribution
